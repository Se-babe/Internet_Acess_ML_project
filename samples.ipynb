```python
# we need to aggregate to household level first by converting individual-level data to household-level data
household_df = df.groupby('hid').agg({
    # household demographics
    'age': 'mean',
    'sex': 'first',

    # education
    'literacy': 'max',
    'Grade': 'max',

    # disability (health dimension)
    'seeing': 'max',
    'hearing': 'max',
    'walking': 'max',
    'remembering': 'max',

    # living standards
    'floor': 'first',
    'roof': 'first',
    'walls': 'first',
    'toilet': 'first',
    'waterdrinking': 'first',
    'energysource': 'first',

    # context
    'rururb': 'first',
    'Region15': 'first'
}).reset_index()

print("Household shape:", household_df.shape)
household_df.head()
```

```python
# -----------------------------
# EDUCATION DEPRIVATION
# -----------------------------
household_df['dep_education'] = (
    (household_df['literacy'] <= 1) |
    (household_df['Grade'] <= 7)
).astype(int)

# -----------------------------
# HEALTH (DISABILITY) DEPRIVATION
# -----------------------------
household_df['dep_health'] = (
    (household_df['seeing'] >= 2) |
    (household_df['hearing'] >= 2) |
    (household_df['walking'] >= 2) |
    (household_df['remembering'] >= 2)
).astype(int)

# -----------------------------
# LIVING STANDARD DEPRIVATIONS
# -----------------------------

# poor floor (example threshold — we can refine later)
household_df['dep_floor'] = (household_df['floor'] <= 11).astype(int)

# unsafe water
household_df['dep_water'] = (household_df['waterdrinking'] >= 14).astype(int)

# poor toilet
household_df['dep_toilet'] = (household_df['toilet'] >= 14).astype(int)

# dirty cooking energy
household_df['dep_energy'] = (household_df['energysource'] >= 15).astype(int)

household_df.head()
```

```python
household_df['mpi_score'] = (
    household_df['dep_education'] +
    household_df['dep_health'] +
    household_df['dep_floor'] +
    household_df['dep_water'] +
    household_df['dep_toilet'] +
    household_df['dep_energy']
) / 6

household_df[['mpi_score']].describe()
```

```python
household_df['poor_MPI'] = (household_df['mpi_score'] >= 0.33).astype(int)

household_df['poor_MPI'].value_counts(normalize=True)
```
